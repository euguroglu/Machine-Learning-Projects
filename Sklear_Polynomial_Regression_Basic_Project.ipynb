## Importing libraries

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import pandas as pd
```

## Creating data

```
np.random.seed(42)
n_samples = 100
X = np.linspace(0,10,100)
error = np.random.rand(n_samples)*100

y = X**3 + error + 100
```

## Data visualization

```
plt.figure(figsize=(12,8))
plt.scatter(X,y)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Linear Regression

```
model = LinearRegression()
model.fit(X.reshape(-1,1),y)
model_pred = model.predict(X.reshape(-1,1))

plt.figure(figsize=(12,8))
plt.scatter(X,y)
plt.plot(X,model_pred,c='r')
print(r2_score(y,model_pred))
```

## Polynomial Regreession

```
from sklearn.preprocessing import PolynomialFeatures
```

```
model_2 = PolynomialFeatures(degree=2)
X_poly = model_2.fit_transform(X.reshape(-1,1))
```

```
model_3 = LinearRegression()
model_3.fit(X_poly,y)
y_pred = model_3.predict(X_poly)
```

```
plt.figure(figsize=(12,8))
plt.scatter(X,y)
plt.plot(X,y_pred,c='r')
plt.xlabel('X data')
plt.ylabel('y data')
plt.title('Visualization')
plt.show()
print(r2_score(y,y_pred))
```
